```python
df['Valor Pagamento']= df['Valor Pagamento'].str.replace(',','.',).astype(float)
```

```python
df['Valor Pagamento'].describe()
```

```python
df.set_index('Valor Pagamento',inplace=True)
```

```python
df.describe()
```

```python
df.reset_index(inplace=True)
```

```python
quantidadeBeneficiados = df['Nome Órgão Superior'].value_counts()
quantidadeBeneficiados
```

```python
valorPorOrgao = df.groupby(['Nome Órgão Superior'])['Valor Pagamento'].sum()
valorPorOrgao
```

```python
valorPorOrgao.plot.bar(title='Valor por orgão')
```

```python
quantidadeBeneficiados.plot.bar(title='Quantidade Beneficiados')
```

```python
newDF= pd.DataFrame([quantidadeBeneficiados])
```

```python
newDF = newDF.T
```

```python
newDF
```

```python
newDF2= pd.DataFrame([valorPorOrgao])
newDF2 = newDF2.T
```

```python
newDF2
```

```python
newDF['Valor Pagamento']=newDF2['Valor Pagamento']
```

```python
newDF['Razao'] = newDF['Valor Pagamento']/newDF['Quantidade Pessoas']
```
